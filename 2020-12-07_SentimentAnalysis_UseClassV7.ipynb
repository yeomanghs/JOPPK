```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
import re

cutSentenceFile = "D:/Users/figohjs/Documents/JOPPK/Data/External/CutSentence.xlsx"
df = pd.read_excel(cutSentenceFile)
cutSentenceDict = df.set_index('Original').to_dict()['Replacement']

def cutSentence(text):
    #cut sentence
    if re.search('|'.join(cutSentenceDict.keys()), text):
        matchList = re.findall('|'.join(cutSentenceDict.keys()), text)
        for match in matchList:
            text = re.sub(match, cutSentenceDict[match], text)     
    return text

#between -1 (most extreme negative) and +1 (most extreme positive)
def giveSentimentScore(text):
    analyzer = SentimentIntensityAnalyzer()
    text = cutSentence(text)
    sentimentscore = analyzer.polarity_scores(text)
    score = sentimentscore['compound']
    return score
```

### 2020-12-07 

```python
import pathlib
import os
from datetime import datetime
import re
import pandas as pd
```

```python
fileList = []
wpList = []
excludedFolder = ['Archive', 'Thumbs.db', 'Earnings template - Shortcut.lnk',
                               '1_Overall Summary', 'Testing']

def extractAllFiles(wp):
    for file in os.listdir(wp):
        if os.path.isfile(wp + '/' + file):
            if '.pdf' in file:
                fileList.append(file)
                wpList.append(wp)
        elif os.path.isdir(wp + '/' + file):
            if file not in excludedFolder:
                extractAllFiles(wp + '/' + file)
```

```python
cdPath = 'D:/Users/figohjs/Documents/JOPPK/Data/Raw/Quarterly Earnings PDF 2018'

extractAllFiles(cdPath)
```

```python
len(wpList)
```

```python
len(fileList)
```

```python
stockNameList = []
dateList = []
periodList = []

for file in zip(wpList, fileList):
    stockName = file[1].split(' ')[0] + '_' + file[1].split(' ')[1]
    stockNameList.append(stockName)
    
    #modified date
    fname = pathlib.Path(file[0] + '/' + file[1])
    dt = datetime.fromtimestamp(fname.stat().st_mtime)
    dateList.append(dt.strftime("%Y-%m-%d"))
    
    if re.search('(.{2} )\d{4}.*', file[1]):
        period = re.search('(.{2} )\d{4}.*', file[1]).group(1)
    else:
        period = ''
    periodList.append(period)
```

```python
df = pd.DataFrame({'Ticker':[re.sub('_', ' ',i) + ' Equity' for i in stockNameList], "Modified Date": dateList,
                 'Period':periodList})
```

```python
periodDict = {'Q1':'1Q', 'H1':'1H', 'H2':'2H','Q3':'3Q', 'Q2':'2Q', 'Q4':'4Q'}
df['Period'] = df['Period'].str.strip().map(lambda x:periodDict[x] if x in periodDict.keys() else x)
```

```python
dfMerged = dfExcel.merge(df, how = 'left', on = ['Ticker', 'Period'])
dfMerged.query("`Modified Date` == ''").shape
```
