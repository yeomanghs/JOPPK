# Version 4

```python
import sentimentScore_v4 as SS

import pandas as pd
```

```python
#2018 and 2019 combined
cdPath = '../data/Raw'

SS_All = SS.SentimentScore(cdPath)

SS_All.extractAllFiles(cdPath)

SS_All.extractAllPDF()

SS_All.processText()

pandaDict_All = SS_All.prepareDict()

pandaDF_All = SS_All.prepareDF()
```

```python
#create dict for date mapping
DateDict = pandaDF_All[['FullStockName', 'Period', 'Year', 'AnnouncementDate']]\
            .set_index(['FullStockName', 'Period', 'Year'])\
            .to_dict()['AnnouncementDate']

#assign overall score
#save to excel
excelFileName = '../Data/Interim/ProcessedTickerFinal_2019-12-12.xlsx'
(pandaDF_All.filter(['FullStockName', 'Period', 'Year', 'ContentType', 'SentimentScore_Magnify'])
            .pivot_table(values = 'SentimentScore_Magnify', 
                         index = ['FullStockName', 'Period', 'Year'], 
                         columns = 'ContentType')
            .reset_index()
            .assign(OverallScore = lambda x:0.4*x['Assessment'] + 
                            0.4*x['Review'] + 0.2*x['Other details'])
            .assign(AnnouncementDate = lambda x:x.apply(lambda y:DateDict[(y['FullStockName'], y['Period'], y['Year'])],
                                                        axis = 1))
            .to_excel(excelFileName, index = False))
```

```python
filename = '../Data/Interim/ProcessedTicker_2019-11-07.xlsx'
df = pd.read_excel(filename)
```

```python
AssessmentScore_Dict = df[['StockName', 'Sector', 'Year', 'Period', 'SentimentScore']]\
                        .set_index(['StockName', 'Sector', 'Year', 'Period'])\
                        .to_dict()['SentimentScore']
```

```python
AssessmentText_Dict = df[['StockName', 'Sector', 'Year', 'Period', 'Content']]\
                        .set_index(['StockName', 'Sector', 'Year', 'Period'])\
                        .to_dict()['Content']
```

```python
filename = '../Data/Interim/ProcessedTicker_2019-11-11.xlsx'
(df.filter(['StockName', 'Sector', 'Year', 'Period', 'SentimentScore'])
   .groupby(['StockName', 'Sector', 'Year', 'Period'])
   .agg({'SentimentScore':'mean'})
   .rename(columns = {'SentimentScore':'AverageScore'})
   .reset_index()
   .assign(AssessmentScore = lambda x:(x.apply(lambda y: AssessmentScore_Dict[(y['StockName'], y['Sector'],
                                                                               y['Year'], y['Period'])], axis = 1)))
   .assign(AssessmentText = lambda x:(x.apply(lambda y: AssessmentText_Dict[(y['StockName'], y['Sector'],
                                                                               y['Year'], y['Period'])], axis = 1))))\
.to_excel(filename, index = False)
```
